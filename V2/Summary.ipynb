# The El Farol Bar Problem
## Our Model
There $A$ agents ($A=100$)  
The bar has capacity for $T$ people to be comfortable ($T = 60$)  
Each agent has $S$ different strategies.


Each strategy predicts attendance as a linear combination of the previous $M$ weeks of attendance.  

$$
\hat{N}^{(i)}(t) = a_1 N(t-1) + a_2 N(t-2) + ... + a_M N(t - M)
$$

Each agent picks the strategy that would have minimized their prediction error over the last $M$ weeks.

Strategy matrix:
$$
\mathbf{S} = \begin{pmatrix}
a_1^{(1)} & a_2^{(1)} & \cdots & a_m^{(1)} \\
a_1^{(2)} & a_2^{(2)} & \cdots & a_m^{(2)} \\
\vdots  & \vdots  & \ddots & \vdots  \\
a_1^{(S)} & a_2^{(S)} & \cdots & a_m^{(S)}
\end{pmatrix}
$$

History matrix:

$$
\mathbf{H} = \begin{pmatrix}
N(t - M -1) & N(t - M) & \cdots & N(t-1) \\
N(t - M -2) & N(t - M-1) & \cdots & N(t-2) \\
\vdots  & \vdots  & \ddots & \vdots  \\
N(t - 2M) & N(t - 2M + 1) & \cdots & N(t - M) 
\end{pmatrix}
$$

Predictions:

$$
\hat{N} = \mathbf{S} \mathbf{H}
$$

The $i$th row of $\hat{N}$ will contain the predictions according to strategy $i$ for the weeks from $t - M$ to $t$

We select the strategy $S^{(*)}$ that minimizes 

$$
\left |
\begin{pmatrix}
\hat{N}(t-1) \\
\hat{N}(t-2) \\
\vdots\\
\hat{N}(t - M)
\end{pmatrix} - \begin{pmatrix}
N(t-1) \\
N(t-2) \\
\vdots\\
N(t - M)
\end{pmatrix}
\right |
$$

and go to the bar if $\hat{N}_{S^{(*)}, t} < 60$
