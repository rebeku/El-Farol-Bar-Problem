```python
import numpy as np
from matplotlib import pyplot as plt

from EFBP import EFBPSim
```

```python
rng = np.random.default_rng(17)

class StrategyMaker:
    def __init__(self, sigma):
        self.sigma = sigma

    def make(self, rng, strategies, memory):
        
        if self.sigma > 0:
            center = rng.normal(0, self.sigma)
            radius = 1 + rng.chisquare(self.sigma)
        else:
            center = 0
            radius = 1
            
        self.min = center - radius
        self.max = center + radius
        return rng.uniform(self.min, self.max, size=(strategies, memory))


sm = StrategyMaker(1)
sm.make(rng, 5, 2)
```

```python
sigma = 0
agents = 10000
s = 2
memory = 2
n_trials=5
n_iter = max(100, memory * 3)
threshold = 60 * agents / 100

h = rng.choice(agents + 1, 2 * memory)
sm = StrategyMaker(sigma)

for _ in range(n_trials):
    sim = EFBPSim(
        memory=memory, 
        strategies=s,
        start=h, 
        n_iter=n_iter,
        agents=agents,
        strategy_func=sm.make,
        threshold=threshold,
        seed=rng.choice(10000))

    plt.figure()
    plt.plot(sim.hist)
    _ = plt.title(f"{sim.hist[-6:]}    s={s}    agents={agents}")
```

```python
sigma = 0
agents = 10000
s = 20
memory = 10
n_trials=5
n_iter = max(100, memory * 3)
threshold = 60 * agents / 100

h = rng.choice(agents + 1, 2 * memory)
sm = StrategyMaker(sigma)

for _ in range(n_trials):
    sim = EFBPSim(
        memory=memory, 
        strategies=s,
        start=h, 
        n_iter=n_iter,
        agents=agents,
        strategy_func=sm.make,
        threshold=threshold,
        seed=rng.choice(10000))

    plt.figure()
    plt.plot(sim.hist)
    _ = plt.title(f"{sim.hist[-6:]}    s={s}    agents={agents}")
```

```python
def find_attendance_std(sigma, s, m):
    agents = 10000
    n_trials=5
    n_iter = max(100, memory * 3)
    threshold = 60 * agents / 100

    h = rng.choice(agents + 1, 2 * memory)
    sm = StrategyMaker(sigma)
    
    stds = []

    for _ in range(n_trials):
        sim = EFBPSim(
            memory=m, 
            strategies=s,
            start=h, 
            n_iter=n_iter,
            agents=agents,
            strategy_func=sm.make,
            threshold=threshold,
            seed=rng.choice(10000))
        stds.append(sim.hist[-6:].std())
        
    return np.array(stds).mean()
```

```python
writefile = "data/heterogeneity.csv"

with open(writefile, "wt", encoding="utf-8") as f:
    f.write("strategies,memory,threshold,sigma,std\n")
```

```python
def track_heterogeneity(strategies, memory, n_trials):
    # store results of each trial
    tmp = np.zeros(n_trials)

    for sigma in np.arange(0,8.5,0.5):
        for k in range(n_trials):
            std = find_attendance_std(sigma, strategies, memory)
            with open(writefile, "a", encoding="utf-8") as f:
                f.write(f"{strategies},{memory},{threshold},{sigma},{std}\n")
```
