# Where are the fixed points?

Suppose there is a fixed point in weekly attendance.  Then at the fixed point, if the history is [N,N,N,N]] and the threshold is $t$, each agent will attend if

$$
\begin{aligned}
\hat{N}(t) &= a_1 N_4 + a_2 N_3 \\
&= N(a_1  + a_2) <t
\end{aligned}
$$

so

$$
a_2 < \frac{t}{N} - a_1
$$

and thus

$$
P_\text{go} = \int_{-1}^1 \int_{-1}^{max(1, \frac{t}{N} - a_1)} f(a_1, a_2) da_2 da_1
$$

where f is the pdf of the $a$ values chosen by each agent. In order for this to be a fixed point, we need

$$
A P_\text{go} = N
$$

If the $a$'s are uniformly distributed between -1 and 1, and  there are $s$ different strategies, we have

$$
f(a_1, a_2) = \frac{s}{4} (1 - \frac{1}{4} \mathcal{I})^{s-1}
$$

where

$$
\mathcal{I} = \begin{cases}
2 |1 - a_1 - a_2| & 0 < |1 - a_1 - a_2| \leq 1 \\
\frac{1}{2} (-|1 - a_1 - a_2|^2 + 6 |1 - a_1 - a_2| - 1) &  1  < |1 - a_1 - a_2| \leq 3
\end{cases}
$$

Solving for $P_\text{go}$ with 
## $s=2$

(see Fixed Point.nb), we define $r = \frac{t}{N}$.  Note that since t and N are strictly positive, r is as well.

$$
P_\text{go} = 
\begin{cases}
 1 & r\geq 2 \\
 \frac{1}{24} \left(-2 r^3+3 r^2+12 r+6\right) & 0<r\leq 1 \\
 \frac{1}{24} \left(2 r^3-15 r^2+36 r-4\right) & 1<r<2
\end{cases}
$$

Since

$$
P_\text{go} = \frac{N}{A} = \frac{t}{rA}
$$

Solving for $r$, (with t = 60, A = 100), we find $r = 0.843971$.  So

$$
N = \frac{t}{r} = 71.095
$$
